# Country Income by Year

```python
income_df = pd.read_csv('../data_files/country_economics_by_year.csv')
income_df.head()
```

```python
# Using pd.melt() to flatten the table. id_vars is the list of columns not to transpose. value_vars is the list of columns that will be transpose. 

income_df = pd.melt(income_df, id_vars=['country_code','country_name'], value_vars=['1990','1991','1992','1993','1994','1995','1996','1997','1998','1999','2000', '2001','2002','2003','2004','2005',
       '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013',
       '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021'], var_name='year', value_name='income_group')
income_df
```

```python
income_df['year'] = pd.to_numeric(income_df['year'], errors='coerce')
income_df.dtypes
```

```python
income_df = income_df.sort_values(by=['country_name', 'year']).reset_index(drop=True)
income_df
```

```python
from pathlib import Path  
filepath = Path('../data_files/country_income_levels.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
income_df.to_csv(filepath) 
```
