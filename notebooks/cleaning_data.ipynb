```python
df3.columns
```

```python
# Reordering the columns in a more sensical way

df4 = df3[['country_name',
           'Year',
           'Total_Population',
           'Urban_Population',
           'GDP',
           'GDP_per_Capita',
           'GDP_Growth',
           'GHG_Growth',
           'GHG_Emissions_per_Capita',
           'Agricultural_Land', 
           'Arable_Land',
           'Forested_Land',
           'CO2_Emissions_by_Sector_Building',
           'CO2_Emissions_by_Sector_Bunker_Fuels',
           'CO2_Emissions_by_Sector_Electricity_Heat',
           'CO2_Emissions_by_Sector_Energy',
           'CO2_Emissions_by_Sector_Fugitive_Emissions',
           'CO2_Emissions_by_Sector_Industrial_Processes',
           'CO2_Emissions_by_Sector_Land_Use_Change_and_Forestry',
           'CO2_Emissions_by_Sector_Manufacturing_Construction',
           'CO2_Emissions_by_Sector_Other_Fuel_Combustion',
           'CO2_Emissions_by_Sector_Total_excluding_LUCF',
           'CO2_Emissions_by_Sector_Total_Including_LUCF',
           'CO2_Emissions_by_Sector_Transportation',
           'Net_Electricity_Consumption',
           'Daily_Coal_Consumption_per_capita',
           'Daily_Gas_Consumption_per_capita',
           'Daily_Oil_Consumption_per_capita']]
```

```python
df4.columns
```

```python
# Rounding the values to 2 decimal places

df5 = df4.round({'Urban Population':2,
            'GDP per Capita':2,
            'GDP Growth':2,
            'GHG Growth':2,
            'GHG Emissions per Capita':2,
            'Agricultural Land':2, 
            'Arable Land':2,
            'Forested Land':2,
            'CO2 Emissions by Sector - Building':2,
            'CO2 Emissions by Sector - Bunker Fuels':2,
            'CO2 Emissions by Sector - Electricity/Heat':2,
            'CO2 Emissions by Sector - Energy':2,
            'CO2 Emissions by Sector - Fugitive Emissions':2,
            'CO2 Emissions by Sector - Industrial Processes':2,
            'CO2 Emissions by Sector - Land Use Change and Forestry':2,
            'CO2 Emissions by Sector - Manufacturing/Construction':2,
            'CO2 Emissions by Sector - Other Fuel Combustion':2,
            'CO2 Emissions by Sector - Total excluding LUCF':2,
            'CO2 Emissions by Sector - Total Including LUCF':2,
            'CO2 Emissions by Sector - Transportation':2,
            'Net Electricity Consumption':2,
            'Daily Coal Consumption per capita':2,
            'Daily Gas Consumption per capita':2,
            'Daily Oil Consumption per capita':2})

```

```python
df5.head()
```

```python
df6 = df5.dropna()
```

```python
df6.isna().sum()
```

```python
# Changing Datatypes 

conversion_dict = {'Year':'int64'}

df7 = df6.astype(conversion_dict)
```

```python
df7.dtypes
```

```python
# Replacing all 0 values with Nan so we can properly drop them if needed and Tableau can recognize the values as missing and not actually 0.

df7.replace(0.00, np.nan, inplace=True)
```
